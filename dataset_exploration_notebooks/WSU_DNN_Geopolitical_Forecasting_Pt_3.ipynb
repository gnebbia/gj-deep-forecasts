```python
import pandas as pd
import numpy as np

ds = pd.read_csv('~/iarpa_geofc.csv', sep=None, engine='python', parse_dates=["date_start","date_fcast_1","date_fcast_2","date_fcast_3"])
```

```python
ds.shape
ds.head()
dso_to_save = ds

ds.head()
```

```python
dso_to_save["recency_2"] = dso_to_save["date_fcast_2"] - dso_to_save["date_start"]
dso_to_save["recency_3"] = dso_to_save["date_fcast_3"] - dso_to_save["date_start"]


```

```python
(dso_to_save["date_fcast_1"].max() - dso_to_save["date_fcast_1"].min()).days 

ds = ds.drop(['date_start','date_fcast_1','date_fcast_2','date_fcast_3'], axis=1)

column_list = ds.columns.tolist()
column_list.remove("recency_1")
column_list.remove("recency_2")
column_list.remove("recency_3")

index_1_c = column_list.index("value_1_c")
column_list = column_list[:index_1_c+1] + ["recency_1"] + column_list[index_1_c+1:]
index_2_c = column_list.index("value_2_c")
column_list = column_list[:index_2_c+1] + ["recency_2"] + column_list[index_2_c+1:]
index_3_c = column_list.index("value_3_c")
column_list = column_list[:index_3_c+1] + ["recency_3"] + column_list[index_3_c+1:]
#column_list.remove(column_list.index("recency_1"))
#column_list.remove(column_list.index("recency_2"))
#column_list.remove(column_list.index("recency_3"))

column_list
```

```python
ds = ds[column_list]
```

```python
ds.shape
```

```python
#ds[ds["new"].dt.days < 0] = pd.Timedelta('1 days')
ds.loc[ds.recency_1.dt.days < 0,'recency_1'] = pd.Timedelta('0 days');
ds.loc[ds.recency_2.dt.days < 0,'recency_2'] = pd.Timedelta('0 days');
ds.loc[ds.recency_3.dt.days < 0,'recency_3'] = pd.Timedelta('0 days');
```

```python
ds["recency_1"]
```

```python
ds = ds.drop(['date_start','date_fcast_1','date_fcast_2','date_fcast_3'], axis=1)
```

```python
ds.to_csv("~/iarpa_geofc_with_recency.csv", index=False) 
```
